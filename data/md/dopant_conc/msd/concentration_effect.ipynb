```python
import matplotlib.pyplot as plt
import numpy as np

# Import MSD files as lists:
ud_step, mg_dil = np.loadtxt('600/mg_dilute_msd_na_redo.txt', skiprows=2, unpack=True)
ud_step, mg_conc = np.loadtxt('600/mg_conc_msd_na.txt', skiprows=2, unpack=True)
ud_step, mg_supconc = np.loadtxt('600/mg_supconc_msd_na.txt', skiprows=2, unpack=True)


steps = ud_step / 1000

mg_dil = mg_dil - mg_dil[1]
mg_conc = mg_conc - mg_conc[1]
mg_supconc = mg_supconc - mg_supconc[1]
    

# Plot MSD: 
plt.plot(steps[1:], mg_dil[1:])
plt.plot(steps[1:], mg_conc[1:])
plt.plot(steps[1:], mg_supconc[1:])
plt.legend(labels = ['undoped (0.1%)', 'Mg dil. (0.3%)', 'Mg con. (1.2%)', 'Mg supconc. (4.2%)'])
plt.xlabel('time, ns', size = 12)
plt.ylabel('MSD, A^2', size = 12)
plt.title('MSD @600K', size=16)

# Find diffusion coefficient:
mg_dil_slope, intercept = np.polyfit(steps[2000:], mg_dil[2000:], 1)
mg_dil_slope_standard_600 = float(mg_dil_slope * (10 ** (-8)))
mg_conc_slope, intercept = np.polyfit(steps[2000:], mg_conc[2000:], 1)
mg_conc_slope_standard_600 = float(mg_conc_slope * (10 ** (-8)))
mg_supconc_slope, intercept = np.polyfit(steps[7400:8000], mg_supconc[7400:8000], 1)
mg_supconc_slope_standard_600 = float(mg_supconc_slope * (10 ** (-8)))
print(f'The diffusion coefficient for the dilute Mg doped structure @600k is {mg_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Mg doped structure @600k is {mg_conc_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the super concentrated Mg doped structure @600k is {mg_supconc_slope_standard_600} m^2/s')#

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())

make_square_axes(plt.gca())

plt.savefig('mg_600.jpg', bbox_inches='tight', dpi=500)
```

```python
d_mg_conc = [mg_conc_slope_standard_600, mg_conc_slope_standard_700, mg_conc_slope_standard_800, mg_conc_slope_standard_900]
d_mg_dil = [mg_dil_slope_standard_600, mg_dil_slope_standard_700, mg_dil_slope_standard_800, mg_dil_slope_standard_900]
d_mg_supconc = [mg_supconc_slope_standard_600, mg_supconc_slope_standard_700, mg_supconc_slope_standard_800, mg_supconc_slope_standard_900]


ts = [600, 700, 800, 900]



# Plot MSD:
plt.plot(ts, d_mg_dil)
plt.plot(ts, d_mg_conc)
plt.plot(ts, d_mg_supconc)
plt.legend(labels = ['Mg dil. (0.3%)', 'Mg con. (1.2%)', 'Mg supconc. (4.2%)'])
plt.xlabel('temperature, T', size = 12)
plt.ylabel('diffusion coefficent m^2 s-1', size = 12)
plt.title('Diffusion coefficients', size=16)
```

```python
c_mg_conc = []
c_mg_dil = []
c_mg_supconc = []


# Conversion of diffusion coefficient to conductivity as follows:
# conducitivity = (charge**2 * charge carrier concentration * diffusion coefficient) / boltzmann constant * temperature
#
# charge is electron charge (sodium singularly charged), so 1.602 * 10**(-19) C
#
# charge carrier concentration calculated as number of Na ions in the simulation cell/cell dimensions
# for undoped C =2.709 * 10 ** 28; dilute divalent doping C = 3.272 * 10**28 atoms/m**3; for concetrated divalent doping C = 3.209 * 10**28 atoms/m**3; for superconcentrated divalent doping C = 3.147 * 10**28 atoms/m**3
#                                  dilute trivalent doping C = 3.276 * 10 **28 atoms/m**3; for concentrated trivalent doping C = 3.229 * 10**28 atoms/m**3
#
# diffusion coefficient calcualted as above fromm MSD plot slopes
#
# boltzmann constant = 1.381 * 10**(-23) J/K
#
# temperature varies with experiment
#
# !!! 10**(-2) added to get final results in S/cm as usually displayed

for t, data in zip(ts, d_mg_conc):
        c_mg_conc.append((((1.602 * 10**(-19))**2 * (3.209 * 10**(28)) * data)/((1.381 * 10**(-23)) * t)) * 10**(-2))
        
for t, data in zip(ts, d_mg_dil):
        c_mg_dil.append((((1.602 * 10**(-19))**2 * (3.272 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_mg_supconc):
        c_mg_supconc.append((((1.602 * 10**(-19))**2 * (3.147 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
# Plot MSD:
plt.plot(ts, c_mg_dil)
plt.plot(ts, c_mg_conc)
plt.plot(ts, c_mg_supconc)
plt.legend(labels = ['Mg dil. (0.3%)', 'Mg con. (1.2%)', 'Mg supconc. (4.2%)'])
plt.xlabel('T, (K)', size = 12)
plt.ylabel('$\mathrm{\sigma}$ (S cm$^\mathrm{-1}$)', size = 12)
plt.title('Conductivity', size=16)

plt.savefig('conductivity.jpg', bbox_inches='tight', dpi=500)
```

```python
# Transform data
import numpy as np

log_mg_conc = []
log_mg_dil = []
log_mg_supconc = []

rec_t = []

t2 = [500, 900]
rec_t2 = []


for data, t in zip(c_mg_conc, ts):
    log_mg_conc.append(np.log10(data * t))
    
for data, t in zip(c_mg_dil, ts):
    log_mg_dil.append(np.log10(data * t))

for data, t in zip(c_mg_supconc, ts):
    log_mg_supconc.append(np.log10(data * t)) 
    
for t in ts:
    rec_t.append(1000/t)
    
for t in t2:
    rec_t2.append(1000/t)
    
# Create figure        
import matplotlib.pyplot as plt
import matplotlib.cm as cmx
col = cmx.get_cmap('Paired')
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, ax = plt.subplots()

# Plot MSD:

ax.scatter(rec_t, log_mg_dil)
linear_model=np.polyfit(rec_t, log_mg_dil, 1)
linear_model_fn=np.poly1d(linear_model)
plt.plot(rec_t2,linear_model_fn(rec_t2))
ax.text(1.6, -2.3, 'Mg, 0.3% (0.77 eV)', color='tab:blue', rotation=-30, size=9)

ax.scatter(rec_t, log_mg_conc)
linear_model=np.polyfit(rec_t, log_mg_conc, 1)
linear_model_fn=np.poly1d(linear_model)
ax.plot(rec_t2,linear_model_fn(rec_t2))
ax.text(1.2, 0, 'Mg, 1.2% (0.58 eV)', color='tab:orange', rotation=-23, size=9)

ax.scatter(rec_t, log_mg_supconc)
linear_model=np.polyfit(rec_t, log_mg_supconc, 1)
linear_model_fn=np.poly1d(linear_model)
plt.plot(rec_t2,linear_model_fn(rec_t2))
ax.text(1.2, -2.5, 'Mg, 4.2% (1.04 eV)', color='tab:green', rotation=-40, size=9)

# Label axes and add secondary axis
ax.set_xlabel('(1000/T) (K$^\mathrm{-1}$)', size = 12)
ax.set_ylabel('log$_\mathrm{10}$[$\mathrm{\sigma}$T, (S cm$^\mathrm{-1}$ K)]', size = 12)

def invert(x):
    # 1/x with special treatment of x == 0
    x = np.array(x).astype(float)
    near_zero = np.isclose(x, 0)
    x[near_zero] = np.inf
    x[~near_zero] = (1 / x[~near_zero])*1000
    return x

secax = ax.secondary_xaxis('top', functions=(invert, invert))
secax.set_xlabel('T (K)', size = 12)

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())

make_square_axes(plt.gca())

plt.savefig('mg_cond.jpg', bbox_inches='tight', dpi=500)

import numpy as np

ln_mg_conc = []
ln_mg_dil = []
ln_mg_supconc = []

rec_t = []


for data, t in zip(c_mg_conc, ts):
    ln_mg_conc.append(np.log(data * 10**2 * t))
    
for data, t in zip(c_mg_dil, ts):
    ln_mg_dil.append(np.log(data * 10**2 * t))
    
for data, t in zip(c_mg_supconc, ts):
    ln_mg_supconc.append(np.log(data * 10**2 * t))

    
for t in ts:
    rec_t.append(1000/t)

    
# Find activation energy:
ln_mg_conc_slope, intercept = np.polyfit(rec_t, ln_mg_conc, 1)
ln_mg_dil_slope, intercept = np.polyfit(rec_t, ln_mg_dil, 1)
ln_mg_supconc_slope, intercept = np.polyfit(rec_t, ln_mg_supconc, 1)

# Conversion of conductivity to activation energy as follows:
# ln(conductivity * T) = -Ea/kT
# Ea = - ln(conductivity * T) * kT
#
# slope = ln(conductivity * T) / (1000/T)
#
# Ea = -1000 * slope * k
#
# Ea (in eV) = -1000 * slope * k * e

ea_mg_conc = -1000 * ln_mg_conc_slope * 1.380649 * 10**(-23) * 6.241509 * 10**18
ea_mg_dil = -1000 * ln_mg_dil_slope * 1.380649 * 10**(-23) * 6.241509 * 10**18
ea_mg_supconc = -1000 * ln_mg_supconc_slope * 1.380649 * 10**(-23) * 6.241509 * 10**18

print('1.23% Mg activation energy:', ea_mg_conc, 'eV')
print('0.27% Mg activation energy:', ea_mg_dil, 'eV')
print('4.2% Mg activation energy:', ea_mg_supconc, 'eV')

plt.savefig('logconductivity_mg.jpg', bbox_inches='tight', dpi=500)
```
