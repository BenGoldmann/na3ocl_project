```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 50)
initial = 1226.84 * np.exp((-x)/0.307)
khandy = 588.38 * np.exp((-x)/0.388)

plt.plot(x, initial, color='blue')
plt.plot(x, khandy, color='red')
plt.xlabel('separation, $\mathrm{\AA}$', size=14)
plt.ylabel('energy, eV', size=14)
#plt.title('Na-O Buckingham optimisation', size=14)
plt.legend(labels=['literature: \n 1226.84e$^\mathrm{-r/0.307}$', 'fitted: \n 588.38e$^\mathrm{-r/0.388}$'], fontsize=12)

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())

make_square_axes(plt.gca())

plt.savefig('nao_fitting.jpg', bbox_inches='tight', dpi=1080)
```
