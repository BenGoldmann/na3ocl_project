# MSD to diffusion coefficient for 600K

```python
import matplotlib.pyplot as plt
import numpy as np

# Import MSD files as lists:
ud_step, ud = np.loadtxt('600k/undoped_msd_na.txt', skiprows=2, unpack=True)
udd_step, ud_dil = np.loadtxt('600k/undoped_dilute_msd_na.txt', skiprows=2, unpack=True)
ud_step, mg_dil = np.loadtxt('600k/mg_dilute_msd_na.txt', skiprows=2, unpack=True)
ud_step, mg_dil_r = np.loadtxt('600k/mg_dilute_msd_na_redo.txt', skiprows=2, unpack=True)
ud_step, mg_conc = np.loadtxt('600k/mg_conc_msd_na.txt', skiprows=2, unpack=True)
ud_step, ca_dil = np.loadtxt('600k/ca_dilute_msd_na.txt', skiprows=2, unpack=True)
ud_step, ca_dil_r = np.loadtxt('600k/ca_dilute_msd_na_redo.txt', skiprows=2, unpack=True)
ud_step, ca_conc = np.loadtxt('600k/ca_conc_msd_na.txt', skiprows=2, unpack=True)
ud_step, al_dil =np.loadtxt('600k/al_dilute_msd_na.txt', skiprows=2, unpack=True)
ud_step, al_conc =np.loadtxt('600k/al_conc_msd_na.txt', skiprows=2, unpack=True)
al1_step, al_dil_r =np.loadtxt('600k/al_dilute_msd_na_redo.txt', skiprows=2, unpack=True)
al2_step, al_conc_r =np.loadtxt('600k/al_conc_msd_na_redo.txt', skiprows=2, unpack=True)
ga1_step, ga_dil =np.loadtxt('600k/ga_dilute_msd_na.txt', skiprows=2, unpack=True)
ga2_step, ga_conc =np.loadtxt('600k/ga_conc_msd_na.txt', skiprows=2, unpack=True)

steps = ud_step / 1000
stepses = udd_step / 1000
r1_steps = al1_step / 1000
r2_steps = al2_step / 1000
g1_steps = ga1_step / 1000
g2_steps = ga2_step / 1000

ud = ud - ud[1]
ud_dil = ud_dil - ud_dil[1]
mg_dil = mg_dil - mg_dil[1]
mg_dil_r = mg_dil_r - mg_dil_r[1]
mg_conc = mg_conc - mg_conc[1]
ca_dil = ca_dil - ca_dil[1]
ca_dil_r = ca_dil_r - ca_dil_r[1]
ca_conc = ca_conc - ca_conc[1]
al_dil = al_dil - al_dil[1]
al_conc = al_conc - al_conc[1]
al_dil_r = al_dil_r - al_dil_r[1]
al_conc_r = al_conc_r - al_conc_r[1]
ga_dil = ga_dil - ga_dil[1]
ga_conc = ga_conc - ga_conc[1]
    

# Plot MSD:   
plt.plot(steps[1:], ud[1:], color='red')
plt.plot(stepses[1:], ud_dil[1:], color='pink')
plt.plot(steps[1:], mg_dil[1:], color='skyblue')
plt.plot(steps[1:], mg_dil_r[1:], color='black')
plt.plot(steps[1:], mg_conc[1:], color='blue')
plt.plot(steps[1:], ca_dil[1:], color='lawngreen')
plt.plot(steps[1:], ca_dil_r[1:], color='grey')
plt.plot(steps[1:], ca_conc[1:], color='green')
plt.plot(steps[1:], al_dil[1:], color='gold')
plt.plot(steps[1:], al_conc[1:], color='darkgoldenrod')
plt.plot(r1_steps[1:], al_dil_r[1:], color='yellow')
plt.plot(r2_steps[1:], al_conc_r[1:], color='brown')
plt.plot(g1_steps[1:], ga_dil[1:], color='pink')
plt.plot(g2_steps[1:], ga_conc[1:], color='pink')
plt.legend(labels = ['undoped (0.27%)', 'undoped dil. (0.08%)', 'Mg dil. (0.27%)', 'Mg dil. (0.27%) R', 'Mg con. (1.23%)', 
                     'Ca dil. (0.27%)', 'Ca dil. (0.27%) R', 'Ca con.(1.23%)', 'Al dil.', 'Al con.', 'Al dil (R)', 'Al conc (R)',
                     'Ga dil', 'Ga conc'])
plt.xlabel('time, ps', size = 12)
plt.ylabel('MSD, A^2', size = 12)
plt.title('MSD @600K', size=16)

# Find diffusion coefficient:
ud_slope, intercept = np.polyfit(steps[2000:], ud[2000:], 1)
ud_slope_standard_600 = float(ud_slope * (10 ** (-8)))
ud_dil_slope, intercept = np.polyfit(stepses[2000:], ud_dil[2000:], 1)
ud_dil_slope_standard_600 = float(ud_dil_slope * (10 ** (-8)))
mg_dil_slope, intercept = np.polyfit(steps[2000:], mg_dil[2000:], 1)
mg_dil_slope_standard_600 = float(mg_dil_slope * (10 ** (-8)))
mg_dil_r_slope, intercept = np.polyfit(steps[2000:], mg_dil_r[2000:], 1)
mg_dil_r_slope_standard_600 = float(mg_dil_r_slope * (10 ** (-8)))
mg_conc_slope, intercept = np.polyfit(steps[2000:], mg_conc[2000:], 1)
mg_conc_slope_standard_600 = float(mg_conc_slope * (10 ** (-8)))
ca_dil_slope, intercept = np.polyfit(steps[2000:], ca_dil[2000:], 1)
ca_dil_slope_standard_600 = float(ca_dil_slope * (10 ** (-8)))
ca_dil_r_slope, intercept = np.polyfit(steps[2000:], ca_dil_r[2000:], 1)
ca_dil_r_slope_standard_600 = float(ca_dil_r_slope * (10 ** (-8)))
ca_conc_slope, intercept = np.polyfit(steps[2000:], ca_conc[2000:], 1)
ca_conc_slope_standard_600 = float(ca_conc_slope * (10 ** (-8)))
al_dil_slope, intercept = np.polyfit(steps[2000:], al_dil[2000:], 1)
al_dil_slope_standard_600 = float(al_dil_slope * (10 ** (-8)))
al_conc_slope, intercept = np.polyfit(steps[2000:], al_conc[2000:], 1)
al_conc_slope_standard_600 = float(al_conc_slope * (10 ** (-8)))
al_dil_r_slope, intercept = np.polyfit(r1_steps[2000:], al_dil_r[2000:], 1)
al_dil_r_slope_standard_600 = float(al_dil_r_slope * (10 ** (-8)))
al_conc_r_slope, intercept = np.polyfit(r2_steps[2000:], al_conc_r[2000:], 1)
al_conc_r_slope_standard_600 = float(al_conc_r_slope * (10 ** (-8)))
ga_dil_slope, intercept = np.polyfit(g1_steps[2000:], ga_dil[2000:], 1)
ga_dil_slope_standard_600 = float(ga_dil_slope * (10 ** (-8)))
ga_conc_slope, intercept = np.polyfit(g2_steps[2000:], ga_conc[2000:], 1)
ga_conc_slope_standard_600 = float(ga_conc_slope * (10 ** (-8)))
print(f'The diffusion coefficient for the undoped structure @600k is {ud_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute undoped structure @600k is {ud_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Mg doped structure @600k is {mg_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Mg doped structure (R) @600k is {mg_dil_r_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Mg doped structure @600k is {mg_conc_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Ca doped structure @600k is {ca_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Ca doped structure (R) @600k is {ca_dil_r_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Ca doped structure @600k is {ca_conc_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Al doped structure @600k is {al_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Al doped structure @600k is {al_conc_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Al doped structure (R) @600k is {al_dil_r_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Al doped structure (R) @600k is {al_conc_r_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the dilute Ga doped structure @600k is {ga_dil_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for the concentrated Ga doped structure @600k is {ga_conc_slope_standard_600} m^2/s')
```

```python
c_ud = []
c_ud_dil = []
c_mg_conc = []
c_mg_dil = []
c_ca_conc = []
c_ca_dil = []
c_al_conc = []
c_al_dil = []
c_ga_conc = []
c_ga_dil = []


# Conversion of diffusion coefficient to conductivity as follows:
# conducitivity = (charge**2 * charge carrier concentration * diffusion coefficient) / boltzmann constant * temperature
#
# charge is electron charge (sodium singularly charged), so 1.602 * 10**(-19) C
#
# charge carrier concentration calculated as number of Na ions in the simulation cell/cell dimensions
# for undoped C =2.709 * 10 ** 28; dilute divalent doping C = 3.272 * 10**28 atoms/m**2; for concetrated divalent doping C = 3.209 * 10**28 atoms/m**2; 
#                                  dilute trivalent doping C = 3.276 * 10 **28 atoms/m**2; for concentrated trivalent doping C = 3.229 * 10**28
#
# diffusion coefficient calcualted as above fromm MSD plot slopes
#
# boltzmann constant = 1.381 * 10**(-23) J/K
#
# temperature varies with dataset
#
# !!! 10**(-2) added to get final results in S/cm as usually displayed

for t, data in zip(ts, d_ud):
        c_ud.append((((1.602 * 10**(-19))**2 * (3.272 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))

for t, data in zip(ts, d_ud_dil):
        c_ud_dil.append((((1.602 * 10**(-19))**2 * (2.709 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))

for t, data in zip(ts, d_mg_conc):
        c_mg_conc.append((((1.602 * 10**(-19))**2 * (3.209 * 10**(28)) * data)/((1.381 * 10**(-23)) * t)) * 10**(-2))
        
for t, data in zip(ts, d_mg_dil):
        c_mg_dil.append((((1.602 * 10**(-19))**2 * (3.272 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_ca_conc):
        c_ca_conc.append((((1.602 * 10**(-19))**2 * (3.209 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_ca_dil):
        c_ca_dil.append((((1.602 * 10**(-19))**2 * (3.272 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
                
for t, data in zip(ts, d_al_conc):
        c_al_conc.append((((1.602 * 10**(-19))**2 * (3.229 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_al_dil):
        c_al_dil.append((((1.602 * 10**(-19))**2 * (3.276 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_ga_conc):
        c_ga_conc.append((((1.602 * 10**(-19))**2 * (3.229 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))
        
for t, data in zip(ts, d_ga_dil):
        c_ga_dil.append((((1.602 * 10**(-19))**2 * (3.276 * 10**(28)) * data)/((1.381 * 10**(-23)) * t) * 10**(-2)))

        
# Plot MSD:
plt.plot(ts, c_ud, color='red')
plt.plot(ts, c_ud_dil, color='pink')
plt.plot(ts, c_mg_conc, color='blue')
plt.plot(ts, c_mg_dil, color='skyblue')
plt.plot(ts, c_ca_conc, color='green')
plt.plot(ts, c_ca_dil, color='lawngreen')
plt.plot(ts, c_al_conc, color='darkgoldenrod')
plt.plot(ts, c_al_dil, color='gold')
plt.legend(labels = ['undoped (0.27%)', 'undoped (0.08%)', 'Mg con. (1.23%)', 'Mg dil. (0.27%)', 'Ca con.(1.23%)', 'Ca dil. (0.27%)', 'Al conc.', 'Al dil.'])
plt.xlabel('T, (K)', size = 12)
plt.ylabel('$\mathrm{\sigma}$ (S cm$^\mathrm{-1}$)', size = 12)
plt.title('Conductivity', size=16)

plt.savefig('conductivity.jpg', bbox_inches='tight', dpi=500)
```
