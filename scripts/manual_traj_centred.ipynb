```python
import matplotlib.pyplot as plt
from matplotlib import patches
import matplotlib.cm as cmx
import matplotlib.patheffects as PathEffects
import numpy as np
import matplotlib as mpl
col = cmx.get_cmap('Paired')
plt.rcParams.update({'mathtext.default':  'regular' })

sf= 0.37
texts = 60
offset = 0.20

lw = 8
olw = 5
dw = 5


radii = {'Na': 1.16*sf, 'O': 1.21* sf, 'Cl': 0.6, 'F': 0.8 * sf, 'M': 1.0* sf}
color = {'Na': 'orange', 'O': '(0.1, 0.2, 0.5)', 'Cl': 'green', 'F': 'cyan', 'M': 'pink'}



def add_patch(ax, x, y, species, vacancy = False):
    r = radii[species]
    c = color[species]
    
    if vacancy:
        p = patches.Rectangle((x - r,y - r), width=r*2, height = r*2, fc = 'white', ec = 'k', linewidth = 1.5)
        ax.add_patch(p)
        t = ax.text(x,y,'$\mathrm{V_{' + species + '}}$', ha = 'center', va = 'center', fontsize = 14)
    else:
        p = patches.Circle((x,y), radius=r, fc = c, ec = 'k', linewidth = lw)
        ax.add_patch(p)
        t = ax.text(x,y,species, ha = 'center', va = 'center', fontsize = texts)
        t.set_path_effects([PathEffects.withStroke(linewidth=olw, foreground='w')])
```
