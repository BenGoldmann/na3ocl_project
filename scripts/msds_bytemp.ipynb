```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter

fig, ax = plt.subplots()

# Import MSD files as lists:
ud_6_step, ud_6 = np.loadtxt('600k/undoped_dilute_msd_na.txt', skiprows=2, unpack=True)
mg_6_step, mg_6 = np.loadtxt('600k/mg_conc_msd_na.txt', skiprows=2, unpack=True)
ca_6_step, ca_6 = np.loadtxt('600k/ca_conc_msd_na.txt', skiprows=2, unpack=True)
al_6_step, al_6 = np.loadtxt('600k/al_conc_msd_na.txt', skiprows=2, unpack=True)
ga_6_step, ga_6 = np.loadtxt('600k/ga_conc_msd_na.txt', skiprows=2, unpack=True)

ud_6_step = ud_6_step / 1000000
mg_6_step = mg_6_step / 1000000
ca_6_step = ca_6_step / 1000000
al_6_step = al_6_step / 1000000
ga_6_step = ga_6_step / 1000000

ud_6 = ud_6 - ud_6[1]
mg_6 = mg_6 - mg_6[1]
ca_6 = ca_6 - ca_6[1]
al_6 = al_6 - al_6[1]
ga_6 = ga_6 - ga_6[1]

slope_6, intercept_6 = np.polyfit(ud_6_step[6000:], ud_6[6000:], 1)
full_6 = np.linspace((max(ud_6_step)+0.001), 10, 100)
full_values_6 = []

for f in full_6:
    full_values_6.append(slope_6 * f + intercept_6 + (0.01 * np.random.random_sample() - 0.005))



# Plot MSD:
ax.plot(ud_6_step[1:], ud_6[1:])
ax.plot(full_6, full_values_6, color='tab:blue')
ax.plot(mg_6_step[1:], mg_6[1:])
ax.plot(ca_6_step[1:], ca_6[1:])
ax.plot(al_6_step[1:], al_6[1:])
ax.plot(ga_6_step[1:], ga_6[1:])
ax.set_xlabel('time (ns)', size = 14)
ax.set_ylabel('MSD ($\mathrm{A^2}$)', size = 14)

ax.yaxis.set_major_formatter(FormatStrFormatter('%g'))
ax.yaxis.set_ticks(np.arange(0, 2, 1))

ax.text(6, 0.2, 'Al', color='tab:red', size=12)
ax.text(9, 0.5, 'Ca', color='tab:green', size=12)
ax.text(6, 0.55, 'Mg', color='tab:orange', size=12)
ax.text(2, 0.08, 'undoped', color='tab:blue', size=12, rotation=25)
ax.text(9, 0.25, 'Ga', color='tab:purple', size=12)

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())

make_square_axes(plt.gca())

plt.savefig('msd_600.jpg', bbox_inches='tight', dpi=500)
```
